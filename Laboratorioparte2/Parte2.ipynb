3. Limpieza y conversión de campos
Convierta columnas como "Height", "Weight", "HP Base", "Attack
Base", "Defense Base", etc., a valores numéricos.
Elimine unidades (“m”, “kg”, “lbs”) y convierta los campos a
tipo float.
Separe las columnas compuestas, por ejemplo:
"Type" → Type1, Type2
"Gender" → Male (%), Female (%)

```python
# Separar la columna 'type' en 'type_1' y 'type_2'
df[['Type_1', 'Type_2']] = df['Type'].str.split(',', n=1, expand=True)

# Eliminar espacios en blanco si los hay
df['Type_1'] = df['Type_1'].str.strip()
df['Type_2'] = df['Type_2'].str.strip()
df = df.drop(columns=['Type'])
df.head()
```

```python
# LIMPIAR Y CONVERTIR HEIGHT 
df['Height_m'] = df['Height'].astype(str).str.extract(r'([0-9]+\.?[0-9]*)')
df['Height_m'] = pd.to_numeric(df['Height_m'], errors='coerce')
df['Height_m'].fillna(df['Height_m'].mean(), inplace=True)
df['Height_m'] = df['Height_m'].round(2)

# LIMPIAR Y CONVERTIR WEIGHT 
df['Weight_kg'] = df['Weight'].astype(str).str.extract(r'([0-9]+\.?[0-9]*)')
df['Weight_kg'] = pd.to_numeric(df['Weight_kg'], errors='coerce')
df['Weight_kg'].fillna(df['Weight_kg'].mean(), inplace=True)
df['Weight_kg'] = df['Weight_kg'].round(2)

# ELIMINAR COLUMNAS ORIGINALES 
df.drop(['Height', 'Weight'], axis=1, inplace=True)



# Verificar tipos y primeros datos

df.head()
```

4. Normalización y verificación final
• Verifique valores nulos y duros:
• Cree un nuevo DataFrame df_clean con los campos limpios
listos para análisis:
["Type1", "Type2", "HP Base", "Attack Base", "Defense
Base", "Speed Base", "Height_m", "Weight_kg", "Base
Exp"]

```python
cols_numericas = [
    "HP Base", "Attack Base", "Defense Base",
    "Speed Base", "Height_m", "Weight_kg", "Base Exp"
]

for col in cols_numericas:
    # Convertir a numérico, forzando errores a NaN
    df[col] = pd.to_numeric(df[col], errors='coerce')
    # Rellenar valores nulos con la media de la columna
    df[col].fillna(df[col].mean(), inplace=True)

# Rellenar las columnas de tipo con valores por defecto
df["Type_1"].fillna(df["Type_1"].mode()[0], inplace=True)
df["Type_2"].fillna("None", inplace=True)

# Crear el nuevo DataFrame limpio listo para análisis
df_clean = df[[
    "Pokemon","Type_1", "Type_2",
    "HP Base", "Attack Base", "Defense Base",
    "Speed Base", "Height_m", "Weight_kg", "Base Exp"
]]
df['Type_2'] = df['Type_2'].astype(str).str.strip().str.lower()

# Reemplaza cualquier variante de none o nan por 0
df['Type_2'] = df['Type_2'].replace(['none', 'nan', 'na', '', 'null'], 0)
df_clean.head(10)
```

Etapa 2: Exploratory Data Analysis (EDA)
1. Análisis estadístico
¿Cuál es el promedio, mínimo y máximo de los atributos base (HP,
Attack, Defense, Speed) de todos los Pokémon?


```python
# Seleccionar las columnas base para el análisis
atributos = ["HP Base", "Attack Base", "Defense Base", "Speed Base"]

# Calcular promedio, mínimo y máximo
estadisticas = df_clean[atributos].agg(["mean", "min", "max"])

print("Estadísticas básicas de los atributos base:")
print(estadisticas)
```

3. Análisis gráfico
Realice un boxplot comparando los valores de Attack Base entre los
tipos principales (Type1).
Identifique qué tipo tiene Pokémon con ataques más altos en promedio.

```python

# Crear un boxplot para comparar Attack Base entre los tipos principales
plt.figure(figsize=(12,6))
sns.boxplot(data=df_clean, x="Type_1", y="Attack Base", palette="Set2")

# Personalizar el gráfico
plt.title("Comparación de Attack Base entre tipos principales (Type_1)")
plt.xlabel("Tipo principal (Type_1)")
plt.ylabel("Attack Base")
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.show()
```

5. Análisis gráfico
Genere un gráfico de barras que muestre la cantidad de Pokémon por
tipo principal (Type1).
¿Qué tipo es el más común?

```python

# Contar cuántos Pokémon hay por tipo principal
conteo_tipos = df_clean["Type_1"].value_counts()

# Crear gráfico de barras
plt.figure(figsize=(10,6))
sns.barplot(x=conteo_tipos.index, y=conteo_tipos.values, palette="viridis")

plt.title("Cantidad de Pokémon por tipo principal (Type_1)", fontsize=14)
plt.xlabel("Tipo principal (Type_1)")
plt.ylabel("Cantidad de Pokémon")
plt.xticks(rotation=45)
plt.show()

# Mostrar el tipo más común
tipo_mas_comun = conteo_tipos.idxmax()
print(f" El tipo más común es: {tipo_mas_comun}")
```

10. Análisis gráfico
Construya un gráfico de violín o boxplot múltiple comparando el
atributo Speed Base entre los tipos Flying, Electric y Ground.
¿Qué tipo de Pokémon tiende a ser más rápido?

```python
# Filtrar solo los tipos deseados
tipos = ["Flying", "Electric", "Ground"]
df_speed = df_clean[df_clean["Type_1"].isin(tipos)]

# Crear el gráfico de violín
plt.figure(figsize=(8, 6))
sns.violinplot(x="Type_1", y="Speed Base", data=df_speed, palette="Set2")

plt.title("Comparación del atributo Speed Base entre tipos Flying, Electric y Ground")
plt.xlabel("Tipo principal (Type_1)")
plt.ylabel("Velocidad base (Speed Base)")
plt.show()
```
